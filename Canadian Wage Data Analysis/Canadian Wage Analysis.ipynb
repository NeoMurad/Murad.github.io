```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = r'C:\Users\Murad\Music\dtdb\SQL projects\Canada_Wage_Data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path, encoding='latin-1')

# Display the DataFrame
print(df)
```

```python
# Split the column based on the '(' character
df[['Occupation', 'NOC']] = df['Occupation (NOC)'].str.rsplit('(', 1, expand=True)

# Drop the original column if needed
df = df.drop('Occupation (NOC)', axis=1)

# Display the result
print(df)
```

```python


# Remove only the parentheses from the 'NOC' values
df['NOC'] = df['NOC'].str.replace(r'[()]', '', regex=True)

# Display the result
print(df)
```

```python
# Define a function to extract the first digit from a string
def extract_first_character(value):
    return value[0]

# Create a new column based on the first digit of 'OldColumn'
df['Occupation_Code'] = df['NOC'].apply(extract_first_character)

# Display the result
print(df)
```

```python
df['Occupation_Code'] = df['Occupation_Code'].astype(int)
df
```

```python
# Define a function to extract the first two digits from a string
def extract_firsttwo_character(value):
    return value[:2]

# Create a new column based on the first digit of 'OldColumn'
df['Occupation_Code_2'] = df['NOC'].apply(extract_firsttwo_character)

# Display the result
print(df)
```

```python
# Define a function to map values to categories

def map_to_category(value):
    if value == 0:
        return "Managers"
    elif value == 1:
        return "Professionals"
    elif value == 2:
        return "Technical Occupations and/or Supervisors"
    elif value == 3:
        return "General Occupations/Trades"
    elif value == 4:
        return "Workers/Machine Operators/Support"
    elif value == 5:
        return "Labourers/Drivers/Support"
    else:
        return "Unknown"
    
df['Position'] = df['Position_Code'].apply(map_to_category)
df
```

```python
#we want to build a hierarchal model to explain the variation of the three wage levels (low, median, high) 
#by two different categories/sets of variables: province/territory level variables and occupation classification level variables



Low = new_df[new_df['Wave_Level'] == 'Low']
Median = new_df[new_df['Wave_Level'] == 'Median']
High = new_df[new_df['Wave_Level'] == 'High']


df_occupation = Median[['OccupationDescription', 'Position']]
df_province =  Median[['Province','mean_temperature', 'Average_house_price', 'bachelors_higher', 'Minimum_wage', 'CPI']]
#df_wage = Median[['Wave_Level', 'Value']]
```

```python
# Assuming 'School_Variable1' and 'School_Variable2' are group-level predictors for 'School'
#X_occupation = df_occupation.groupby('Occupation')[['OccupationDescription', 'OccupationDescription_2', 'Position']].first()

# Assuming 'Class_Variable1' and 'Class_Variable2' are group-level predictors for 'Class'
#X_province = df_province.groupby('Province')[['mean_temperature', 'Average_house_price', 'bachelors_higher', 'Minimum_wage', 'CPI']].first()





# columns to keep
columns_to_keep = ['Province', 'Occupation']

for column in Median.select_dtypes(include='object').columns:
    Median[column] = pd.Categorical(Median[column])

# keeping only specificed columns
Median_grouping = Median[columns_to_keep]

categorical_columns = Median_grouping.select_dtypes(include='category').columns

for column in categorical_columns:
    dummy_columns = pd.get_dummies(Median_grouping[column], prefix=column, drop_first=True)
    Median_grouping = pd.concat([Median_grouping, dummy_columns], axis=1)
    Median_grouping = Median_grouping.drop(column, axis=1)
    
```

```python
for column in df_occupation.select_dtypes(include='object').columns:
    df_occupation[column] = pd.Categorical(df_occupation[column])
    
    
for column in df_province.select_dtypes(include='object').columns:
    df_province[column] = pd.Categorical(df_province[column])
    


categorical_columns = df_occupation.select_dtypes(include='category').columns

for column in categorical_columns:
    dummy_columns = pd.get_dummies(df_occupation[column], prefix=column, drop_first=True)
    df_occupation = pd.concat([df_occupation, dummy_columns], axis=1)
    df_occupation = df_occupation.drop(column, axis=1)
    
    
categorical_columns = df_province.select_dtypes(include='category').columns

for column in categorical_columns:
    dummy_columns = pd.get_dummies(df_province[column], prefix=column, drop_first=True)
    df_province = pd.concat([df_province, dummy_columns], axis=1)
    df_province = df_province.drop(column, axis=1)
    
    




```

```python
import statsmodels.api as sm
import pandas as pd



# Combine variables at both levels
X = pd.concat([df_occupation, df_province], axis=1)

# Add a constant term for the intercept
X = sm.add_constant(X)
X = X.dropna()

# Define the hierarchical structure
#grouping = Median[['Occupation', 'Province']]


# Dependent variable
y = Median['Value']

# Fit the Fixed Model
model = sm.OLS(y, X)
result = model.fit()

# Display results
print(result.summary())
```
